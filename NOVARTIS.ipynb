# Stationarity Testing
Most of the TS models work on the assumption that the TS is stationary. Intuitively, we can sat that if a TS has a particular behaviour over time, there is a very high probability that it will follow the same in the future.

```python
import pandas as pd
import numpy as np
```

## ADF(Augmented Dickey-Fuller test):
This is one of the statistical tests for checking stationarity. Here the null hypothesis is that the TS is non-stationary. The test results comprise of a Test Statistic and some Critical Values for difference confidence levels. If the ‘Test Statistic’ is less than the ‘Critical Value’, we can reject the null hypothesis and say that the series is stationary. 

```python
from statsmodels.tsa.stattools import adfuller
```

## Hourly sales

```python
data = pd.read_csv('hourlysales.csv')
data.head()
```

```python
def adfTest(name):
    result = adfuller(name)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
```

```python
adfTest(data.M01AB)
```

```python
adfTest(data.M01AE)
```

```python
adfTest(data.N02BA)
```

```python
adfTest(data.N02BE)
```

```python
adfTest(data.N05B)
```

```python
adfTest(data.N05C)
```

```python
adfTest(data.R03)
```

```python
adfTest(data.R06)
```

### For all the drugs ADF Statistic is less than the critical values and p value <= 0.05 so the drugs are stationary

# FORECASTING METHODS

## Daliy sales

```python
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error
from math import sqrt
```

## M01AB

```python
df = pd.read_csv('dailysales.csv', usecols=['datum', 'M01AB'])
df.head()
#Printing head
#Creating train and test set 
#Index 10392 marks the end of October 2013 
train=df[0:1800] 
test=df[1800:]

#Aggregating the dataset at daily level
df['Timestamp'] = pd.to_datetime(df.datum,format='%d-%m-%Y') 
df.index = df.Timestamp 
df = df.resample('D').mean()
train['Timestamp'] = pd.to_datetime(train.datum,format='%d-%m-%Y') 
train.index = train.Timestamp 
train = train.resample('D').mean() 
test['Timestamp'] = pd.to_datetime(test.datum,format='%d-%m-%Y') 
test.index = test.Timestamp 
test = test.resample('D').mean()
```

### 1) Naive Approach

```python
dd= np.asarray(train["M01AB"])
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
plt.figure(figsize=(12,8))
plt.plot(train.index, train['M01AB'], label='Train')
plt.plot(test.index,test['M01AB'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
plt.savefig('time-series-naive-approach.png')

rms = sqrt(mean_squared_error(test.M01AB, y_hat.naive))
print("RMS value with naive approach : ", rms)
```

### 2) Simple Average 

```python
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['M01AB'].mean()
plt.figure(figsize=(12,8))
plt.plot(train['M01AB'], label='Train')
plt.plot(test['M01AB'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(test.M01AB, y_hat_avg.avg_forecast))
print("RMS value with simple average : ", rms)
```

### 3) Moving Average

```python
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['M01AB'].rolling(60).mean().iloc[-1]
plt.figure(figsize=(16,8))
plt.plot(train['M01AB'], label='Train')
plt.plot(test['M01AB'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(test.M01AB, y_hat_avg.moving_avg_forecast))
print("RMS value with moving average : ", rms)
```

### 4) Exponential smoothing

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['M01AB'])).fit(smoothing_level=0.6,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot(train['M01AB'], label='Train')
plt.plot(test['M01AB'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(test.M01AB, y_hat_avg.SES))
print("RMS with Exponential smoothing", rms)
```

### 5) Holt's linear trend method

```python
y_hat_avg = test.copy()

fit1 = Holt(np.asarray(train['M01AB'])).fit(smoothing_level = 0.3,smoothing_slope = 0.1)
y_hat_avg['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(16,8))
plt.plot(train['M01AB'], label='Train')
plt.plot(test['M01AB'], label='Test')
plt.plot(y_hat_avg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(test.M01AB, y_hat_avg.Holt_linear))
print("RMS value with Holt's linear trend method : ", rms)
```

### 5) Holt winters method

```python
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['M01AB']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['M01AB'], label='Train')
plt.plot(test['M01AB'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(test.M01AB, y_hat_avg.Holt_Winter))
print("RMS value with Holt winters method : ", rms)
```

#### Efficency of model depends on RMS values. Model with less RMS value is more suitable  
#### So, Moving average method is more suitable for M01AB
